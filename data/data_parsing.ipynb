```python
import pandas as pd
```

```python
big_df = pd.read_csv("all_data.csv")
```

```python
big_df
```

```python
big_df = big_df[["name", "season year", "team", "goals", "assists", "totalShots", "minutesPlayed"]]
```

```python
big_df.dropna() # Get rid of Null values
```

```python
big_df = big_df[big_df['season year'] == '15/16']
```

```python
big_df.reset_index()
```

```python
big_df.columns
```

```python
# Rename columns
big_df.columns = ["Name", "Year", "Club", "Goals", "Assists", "Shots", "Minutes Played"]
```

```python
big_df
```

```python
################ IMPORT NATIONALITY and Position Dataset ########################
```
